# TRACKING DES DONNEES ENERGETIQUES EN ROUMANIE DE JANVIER 2019 à MARS 2023.

Cette étude permet d'analyser et visualiser les tendances de la consommation d'énergie ainsi que de la production d'énergie en Roumanie en ce qui concerne les différents types de production d'énergie des sources d'énergie renouvelables et non renouvelables.

Plus bas, la consommation d'énergie par rapport à la production d'énergie est également visualisée en regroupant les données données en mois et en jour de la semaine.

J'espère que vous apprécierez cette analyse qui est transposable à tous les pays d'Europe à partir d'un receuil de données énergétiques sur des sites fiables. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Lecture des données

```python
df = pd.read_csv('electricityConsumptionAndProductioction.csv')
```

# Analyse Exploratoire des données

```python
df.head()
```

```python
df.rename(columns={"Consumption":"Consommation",
                   "Nuclear":"Nucleaire",
                   "Wind":"Eolienne",
                   "Oil and Gas":"Pétrole_&_Gaz",
                  "Coal":"Charbon",
                   "Solar":"Solaire",
                   "Biomass":"Biomasse"},inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df.set_index('DateTime')
df.index=pd.to_datetime(df.index)
```

```python
df.shape
```

# Graphe de la consommation d'énergie individuelle et de la production d'énergie  en fonction du temps.

Tracés en graphes séparés:

Consommation electrique dans le temps 

Production electrique dans le temps

Production et consommation dans le temps

```python
plt.figure(figsize=(20,6))
plt.plot(df.index, df['Consommation'], 'r')
plt.title("Distribution de la Consommation d'energie en fonction du temps")
plt.xlabel('Date')
plt.ylabel('MW')
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.plot(df.index, df['Production'], 'g')
plt.title("Distribution de la Production d'énergie en fonction du temps")
plt.xlabel('Date')
plt.ylabel('MW')
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.plot(df.index, df['Consommation'], 'b')
plt.plot(df.index, df['Production'], 'y')
plt.title("Superposition de la distribution de la Production et Consommation Electrique dans le temps")
plt.xlabel('Date')
plt.ylabel('MW')
plt.show()
```

# Distribution  de chaque type de production d'électricité en Roumanie
#### Cela montrera la tendance de chaque type de source d'énergie. Cela montrera également comment les sources d'énergie renouvelables comme l'énergie solaire, l'énergie hydroélectrique et le biogaz jouent aujourd'hui un grand rôle.

```python
Types = list(df.columns)
Types.pop(0)
Types.pop(0)
Types
```

```python
colors = {'Nuclear':'orange', 'Wind':'aquamarine', 'Hydroelectric':'b', 'Oil and Gas':'saddlebrown', 'Coal':'black', 'Solar':'y', 'Biomass':'crimson'}
```

```python
for prType in Types:
    plt.figure(figsize=(16,5))
    plt.plot(df.index, df[prType], color=colors.get(prType))
    plt.title(f'Distribution de la  production {prType}  en fct du temps')
    plt.xlabel('Date')
    plt.ylabel('MW')
    plt.show()
```

# Création d'attributs DateTime individuels tels que l'année, le mois, le jour et la semaine du jour
#### Ceci est utilisé pour que les données puissent être regroupées pour une analyse plus approfondie.

```python
df['Day'] = df.index.day
df['Year'] = df.index.year
df['Month'] = df.index.month
df['WeekDay'] = df.index.day_name()
```

## Consommation électrique par jour de la semaine
#### Ce graphique montre la consommation d'énergie pour la production pour chaque semaine de la journée

```python
dfbyDay = df.groupby('WeekDay').mean()
```

```python
plt.figure(figsize=(8,4))
plt.plot(dfbyDay.index, dfbyDay['Consommation'], 'dodgerblue')
plt.plot(dfbyDay.index, dfbyDay['Production'], 'r')
plt.xlabel("Consommation et Production d'Energie par jour de semaine")
plt.ylabel('MW')
plt.legend(['Consumption', 'Production'])
plt.show()
```

# Consommation électrique par mois
#### Analyse importante car la consommation d'énergie dépend du climat et de la saison. Ainsi, tracer cela peut aider à comprendre quand baisser ou augmenter la production d'énergie.

```python
dfbyMonth = df.groupby('Month').mean()
```

```python
plt.figure(figsize=(12,4))
plt.plot(dfbyMonth.index, dfbyMonth['Consommation'], 'dodgerblue')
plt.plot(dfbyMonth.index, dfbyMonth['Production'], 'r')
plt.xlabel("Consommation et Production d'Energie par Mois en Roumanie")
plt.ylabel('MW')
plt.legend(['Consommation', 'Production'])
plt.show()
```
